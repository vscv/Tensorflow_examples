```python
def get_points_dict(CSV_name):
    points_dict={}
    with open(CSV_name, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        count=0
        for row in csvreader:
            if count==0:
                count+=1
                continue
            points_dict[row[0]]= [int(x) for x in row[1:]]
            
    return points_dict     
```
