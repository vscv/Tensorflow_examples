```python
mpl.rcParams['lines.linewidth'] = 1
mpl.rcParams['axes.prop_cycle'] = cycler(color=['r', 'g', 'b', 'y'])



fig, ax1 = plt.subplots(figsize=(25, 10))


color = 'tab:red'
ax1.set_title('K-model-TL val_accuracy with lr', fontsize='xx-large')
ax1.set_xlabel('epoch')
ax1.set_ylabel('val_accuracy', color=color)

for cyc in range(8):
    data = np.random.randn(50)
    plt.plot(data, label=str(cyc))
    
ax1.legend(title='Nets:', title_fontsize='x-large', fontsize='large')
```

## Styling with cycler colors

https://matplotlib.org/stable/tutorials/intermediate/color_cycle.html


```python
import numpy as np
import matplotlib.pyplot as plt
from cycler import cycler
```

```python
x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 4, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
```

```python
print(yy.shape)
```

```python
default_cycler = (cycler(color=['r', 'g', 'b', 'y']) +
                  cycler(linestyle=['-', '--', ':', '-.']))

plt.rc('lines', linewidth=4)
plt.rc('axes', prop_cycle=default_cycler)
```

```python

custom_cycler = (cycler(color=['c', 'm', 'y', 'k']) +
                 cycler(lw=[1, 2, 3, 4]))

fig, (ax0, ax1) = plt.subplots(nrows=2, figsize=(25, 10))

# default is default_cycler
ax0.plot(yy)
ax0.set_title('Set default color cycle to rgby')

ax1.set_prop_cycle(custom_cycler) # set to use custom_cycler
ax1.plot(yy)
ax1.set_title('Set axes color cycle to cmyk')

# Add a bit more space between the two plots.
fig.subplots_adjust(hspace=0.3)
plt.show()
```

## 24 cycler for leaf benchmaker

```python
x = np.linspace(0, 2 * np.pi, 50)
offsets = np.linspace(0, 2 * np.pi, 24, endpoint=False)
yy = np.transpose([np.sin(x + phi) for phi in offsets])
```

```python
print(yy.shape)
```

```python
auto_custom_cycler_01 = (cycler(color=[plt.get_cmap('jet')(i/13) for i in range(24)]) +
                  cycler(linestyle=['-', '--', ':', '-.']*6))

fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.set_prop_cycle(auto_custom_cycler_01) # set to use custom_cycler
ax1.plot(yy)
ax1.set_title('Set axes color cycle to cmyk')

# Add a bit more space between the two plots.
fig.subplots_adjust(hspace=0.3)
plt.show()
```

```python


auto_custom_cycler_01 = (cycler(color=[plt.get_cmap('jet')(i/13) for i in range(24)]) +
                  cycler(linestyle=['-', '--', ':', '-.'] * 6)) # [4]*6 = 24

fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.set_prop_cycle(auto_custom_cycler_01) # set to use custom_cycler
for cyc in range(24):
    plt.plot(yy[cyc], label=str(cyc))
    
ax1.legend(title='Nets:', title_fontsize='x-large', fontsize='large')
ax1.set_title('Set axes color cycle to cmyk')



# Add a bit more space between the two plots.
fig.subplots_adjust(hspace=0.3)
plt.show()


```

```python
mpl.rcParams['lines.linewidth'] = 1.5
mpl.rcParams["markers.fillstyle"] = 'left' # 'full', 'left', 'right', 'bottom', 'top', 'none'


auto_custom_cycler_01 = (cycler(color=[plt.get_cmap('jet')(i/13) for i in range(24)]) + # 24 colors
                  cycler(linestyle=['-', '--', ':', '-.'] * 6) + # [4]*6 = 24
                    cycler(marker=['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D'] * 2)) # [12]*2 = 24

fig, ax1 = plt.subplots(figsize=(20, 9))


ax1.set_prop_cycle(auto_custom_cycler_01) # set to use custom_cycler
for cyc in range(24):
    plt.plot(yy[cyc], label=str(cyc))
    
ax1.legend(title='Nets:', title_fontsize='x-large', fontsize='large', bbox_to_anchor=(1.15, 1.0))
ax1.set_title('Set axes color cycle to cmyk')



# Add a bit more space between the two plots.
fig.subplots_adjust(hspace=0.3)
plt.show()
```
